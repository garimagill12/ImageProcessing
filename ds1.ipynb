python provides extensive libraries that can handle such multi dimensional
data. 4 primary libraries that are populor for handling, loading and reading of
images are
  1. pillow

  4. openCV

if we want to load images in czi format especially if we are working on
microscopic images then we can use the library Czifile.

OME-TIFF images( open microscopy enviornment tiff format) - which is a
tiff file with embedded XML which has all the metadata about that specific image.
pyometiff is a Python library for reading and writing OME-TIFF files

Bio-Formats is a software tool for reading and writing image data using
standardized, open formats.their recommended file format is OME-TIFF file
as we do not loose the metadata corr. to images in this format

metadata is important for eg. an img is 20 pixels, the representation of these
pixels ie. if 1 pixel = 100km(sattelite imagery) or 1 pixel = 1 micron(microscopic
 unit)
